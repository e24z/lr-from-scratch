```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('./sales-data.csv')
print(data)
```

```python
x = data['TV']
y = data['Sales']

plt.xlabel('TV Advertising Budget ($)')
plt.ylabel('Sales')
plt.scatter(x, y)
```

Scatter graph seems to describe a linear relationship for the data, so we use a linear model for predictions on this dataset.
